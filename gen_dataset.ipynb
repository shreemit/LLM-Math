```python
import random
import pandas as pd

# Define the size of the dataset
dataset_size = 10000

# Define the range of numbers for operations
range_start = 999
range_end = 10000

# Define the formats
formats = {
    "addition": [
        "what is the addition of {} + {}",
        "what is {} plus {}",
        "find the sum of {} and {}",
        "{} + {}",
        "give me the addition of {} and {}",
        "calculate {} + {}",
    ],
    "subtraction": [
        "{} - {}",
        "what is {} - {}",
        "what is {} minus {}",
        "subtract {} from {}",
        "what is the subtraction of {} - {}",
        "calculate {} - {}",
        "give me the subtraction of {} and {}",
        "tell me {} minus {}",
    ],
    "multiplication": [
        "what is the multiplication of {} * {}",
        "what is {} times {}",
        "find the product of {} and {}",
        "{} times {}",
        "give me the multiplication of {} and {}",
        "calculate {} * {}",
        "give me the product of {} and {}",
        "can you tell me {} times {}",
    ],
    "division": [
        "what is the division of {} / {}",
        "what is {} divided by {}",
        "find the quotient of {} and {}",
        "{} / {}",
        "give me the division of {} and {}",
        "calculate {} / {}",
        "give me the quotient of {} and {}",
    ],
}

# Generate the dataset
data = {
    'Number1': [],
    'Number2': [],
    'Operation': [],
    'Operator': [],
    'Result': [],
}

for _ in range(dataset_size):
    operation = random.choice(list(formats.keys()))
    num1 = random.randint(range_start, range_end)
    num2 = random.randint(range_start, range_end)
    if operation == "division" and num2 == 0:
        num2 = 1  # Avoid division by zero
    format_str = random.choice(formats[operation])
    data['Number1'].append(num1)
    data['Number2'].append(num2)
    data['Operator'].append(operation)
    data['Operation'].append(format_str.format(num1, num2))
    if operation == "addition":
        data['Result'].append(int(num1 + num2))
    elif operation == "subtraction":
        data['Result'].append(int(num1 - num2))
    elif operation == "multiplication":
        data['Result'].append(int(num1 * num2))
    elif operation == "division":
        data['Result'].append(num1 / num2)

# Convert the dataset to a DataFrame
df1 = pd.DataFrame(data)

# Print the DataFrame
df1
```

```python
import random
import pandas as pd
import numpy as np

# Define the size of the dataset
dataset_size = 9000

# Define the range of numbers for matrix elements
range_start = 1
range_end = 10

# Define the formats
formats = {
    "matrix_multiplication": [
        "Multiply the following matrices:\n{}\nand\n{}",
        "What is the product of these matrices?\n{}\nand\n{}",
        "Find the matrix product of:\n{}\nand\n{}",
    ],
    "eigenvalue": [
        "Find the eigenvalues of this matrix:\n{}",
        "What are the eigenvalues of the following matrix?\n{}",
        "Calculate the eigenvalues for this matrix:\n{}",
    ],
    "determinant": [
        "Find the determinant of this matrix:\n{}",
        "What is the determinant of the following matrix?\n{}",
        "Calculate the determinant for this matrix:\n{}",
    ],
}

# Generate the dataset
data = {
    'Matrix1': [],
    'Matrix2': [],
    'Operation': [],
    'Operator': [],
    'Result': [],
}

for _ in range(dataset_size):
    operation = random.choice(list(formats.keys()))
    matrix1 = np.random.randint(range_start, range_end, size=(5, 5))
    matrix2 = np.random.randint(range_start, range_end, size=(5, 5)) if operation == "matrix_multiplication" else None
    format_str = random.choice(formats[operation])
    data['Matrix1'].append(matrix1)
    data['Matrix2'].append(matrix2 if matrix2 is not None else '-')
    data['Operator'].append(operation)
    data['Operation'].append(format_str.format(matrix1, matrix2) if matrix2 is not None else format_str.format(matrix1))
    if operation == "matrix_multiplication":
        data['Result'].append(np.dot(matrix1, matrix2))
    elif operation == "eigenvalue":
        data['Result'].append(np.linalg.eigvals(matrix1))
    elif operation == "determinant":
        data['Result'].append(np.linalg.det(matrix1))

# Convert the dataset to a DataFrame
df2= pd.DataFrame(data)

# Print the DataFrame
df2
```
